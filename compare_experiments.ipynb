```python
out_dirs= ['exp_attenuation_10000.0', 'exp_attenuation_100000.0']
```

```python
import os
import pickle
import matplotlib.pyplot as plt
```

```python
data = {}
for out_dir in out_dirs:
    log_file = os.path.join(out_dir, "log.pkl")
    with open(log_file, "rb") as fp:
        data[out_dir] = pickle.load(fp)
```

```python
x_label = "iter"
y_label = "val/loss"
def plot(x_label, y_label):
    for name in data:
        x = [d[x_label] for d in data[name]]
        y = [d[y_label] for d in data[name]]
        plt.plot(x, y, label = name)
    plt.legend()
    plt.xlabel(x_label)
    _ = plt.ylabel(y_label)
plot("iter", "lr")
```

```python
plot("iter", "val/loss")
```

```python
plot("iter", "train/loss")
```
