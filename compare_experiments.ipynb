```python
#out_dirs= ['exp_attenuation_10000.0', 'exp_attenuation_100000.0']
#out_dirs = ["out-shakespeare-char"]
out_dirs= ['exp_attenuation_10000.0', 'exp_attenuation_100000.0', 'exp_attenuation_1000000.0']
```

```python
import os
import pickle
import matplotlib.pyplot as plt
```

```python
data = {}
for out_dir in out_dirs:
    log_file = os.path.join(out_dir, "log.pkl")
    with open(log_file, "rb") as fp:
        data[out_dir] = pickle.load(fp)
```

```python
list(data[out_dirs[0]][0].keys())
```

```python
x_label = "iter"
y_label = "val/loss"
def plot(x_label, y_label):
    for name in data:
        x = [d[x_label] for d in data[name]]
        y = [d[y_label] for d in data[name]]
        plt.plot(x, y, label = name)
    plt.legend()
    plt.xlabel(x_label)
    _ = plt.ylabel(y_label)
plot("iter", "lr")
```

```python
def plots(x_label, y_labels):
    ncols = len(y_labels)
    fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(15,5))
    for i in range(ncols):
        y_label = y_labels[i]
        for name in data:
            x = [d[x_label] for d in data[name]]
            y = [d[y_label] for d in data[name]]
            axes[i].plot(x, y, label = name)
        axes[i].legend()
        axes[i].set_xlabel(x_label)
        axes[i].set_ylabel(y_label)
plots("iter", ["train/loss", "val/loss"])
```
